## Business Problem

Unsung real estate agency is an agency in King County that deals with clients who want to buy/sell homes.

They want to create an algorithm and use it to make a model for pricing of houses:
- The primary purpose of the model is to predict buying/selling prices of houses.
- The secondary purpose is for inferring features/atrributes that significantly affect the pricing of a house. 

By completing this, the business is at a better position to advise clients who want to buy/sell homes with factual information about homes, their costs or selling prices and exactly why their prices are as they are.

## Overview

## Data Summary

The data used is for the for the year 2014 May - 2015 May. The data has no missing records or inappropriate datatypes. There are 21597 records of homes in King County, their attributes, renovation year if any, and their selling price. This is significant information to use for modelling and inferencing in line with  our business problem.

For our model:
- The target is __`price`__ which is the pricing/value of a house.
- `sqft_lot` is the area of the whole plot of land.
- `sqft_living` is the area which the house/home takes in the plot.
- `floors` are the levels in a house.
- `waterfront` is whether a house has a view of a waterfront or not.
- `bedrooms`and `bathrooms` are the number of the respective rooms in a house.
- `condition` is the overall condition of a house.
- `grade` is the grading of a house.

Plot histograms for all numeric columns

```python
def Hist_plot(data, column):
    """
    A function that does a histogram for the column specified in a given dataframe
    """
    # Set number of bins depending on max value in the column
    if data[column].max() < 1000:
        # Plot hist without bins specifications
        fig, ax = plt.subplots(figsize=(8, 5))    
        ax.hist(data[column])
        
        ax.set_xlabel(column)
        ax.set_ylabel('Count')
        ax.set_title(f'{column.title()} histogram')
        plt.savefig(f'/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/{column.title()} histogram.jpg')
        plt.show()
        
    elif data[column].max() < 1000:
        # Plot hist with 100 bins 
        fig, ax = plt.subplots(figsize=(8, 5))    
        ax.hist(data[column], bins=100)
        
        ax.set_xlabel(column)
        ax.set_ylabel('Count') 
        ax.set_title(f'{column.title()} histogram')
        plt.savefig(f'/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/{column.title()} histogram.jpg')
        plt.show()
        
    else:
        # Plot hist with 200 bins 
        fig, ax = plt.subplots(figsize=(8, 5))    
        ax.hist(data[column], bins=100, align='mid')
        
        ax.set_xlabel(column)
        ax.set_ylabel('Count')
        ax.set_title(f'{column.title()} histogram')
        plt.savefig(f'/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/{column.title()} histogram.jpg')
        plt.show()
                      
```

```python
# Make a list of columns that don't make sense looking for outliers 
# also price as its too early to deal with it
irr_cols = ['long', 'lat', 'zipcode', 'id']

# Create a for loop to loop through our function for plotting
for column in data.columns:
    if data[column].dtype == 'int64' or data[column].dtype == 'float64':
        if column not in irr_cols:
            # call the function
            Hist_plot(data, column)
        
    else:
        continue
```

#### Creating, editing and dropping columns

```python
def Drop_columns(data, columns):
    "This function takes in a list of columns that you want to drop from a dataframe and the dataframe itself"
    data.drop(columns=columns, inplace=True)
    
```

##### Checking columns that could contain information about homes in a neighbourhood

Compare to the sqft_living15 and sqft_lot15 columns in our data as these columns have something to do with neighbourhoods.

```python
data[['sqft_living', 'sqft_living15', 'sqft_lot', 'sqft_lot15']]
```

I can't make sense of sqft_living15 and sqft_lot15 columns so I am going to drop them.

```python
Drop_columns(data, ['sqft_lot15', 'sqft_living15'])
```

##### Create month_sold and year_sold  columns

The date column contains dates that homes were bought and wouldn't be of much help as it is. The column would be better represented as: 
- months of the year(month_sold) and 
- year(year_sold) 

The middle set of numbers in the date records represent month while the first set represent the year.

```python
data.date.head()
```

```python
# Selecting the month indicator in date column
data['month_sold'] = [date[5:7] for date in data['date']]
data['year_sold'] = [date[0:4] for date in data['date']]

# Convert month data type to int
data.month_sold = data.month_sold.astype(int)
data.year_sold = data.year_sold.astype(int)

# Drop date column
Drop_columns(data, ['date'])
```

```python
data.info()
```

##### Create selling_age column

This is the age of a house at the time it is being sold.

    selling_age = year_sold - yr_built

```python
data['selling_age'] = data['year_sold'] - data['yr_built']
```

##### Create renovation_age column

This is the age of a house when the renovation is done.

    renovation_age = yr_renovated - yr_built

```python
data['renovation_age'] = data['yr_renovated'] - data['yr_built']
data['renovation_age'].head()
```

Take care of negative values in the new column. Those records are the houses that have no renovations. Set them as 0 indicating no renovation has been done.

```python
# Replace all negative values with 0
data['renovation_age'] = [y if y > 0 else 0 for y in data['renovation_age']]
```

##### Create sold_renovation column

This is the age in years a house is being sold after renovation.

    sold_renovation = year_sold - yr_renovated

```python
data['sold_renovation'] = data['year_sold'] - data['yr_renovated']
```

```python
data['sold_renovation']
```

Take care of the records that didn't have any renovations done

```python
# Replace all values greater than 1000
# Values past 1000 are for the homes that have no renovations done
data['sold_renovation'] = [y if y < 1000 else 0 for y in data['sold_renovation']]
```

```python
data['sold_renovation']
```

```python
data.head()
```

### Bivariate Analysis

##### Price against all other columns

The main focus in this section will be relationships of other columns in the data to price

```python
X_data = data.drop(columns='price')
y_data = data['price']
```

Check out the column correlations to price:

```python
# Empty list to store correlation values
corr_list = []

for col in X_data.columns:
    corr = data[[y_data.name, col]].corr(method='pearson')[y_data.name][1]
    corr_list.append(round(corr, 5))
    
# Matching corr values to a column name
column_corr = list(zip(X_data.columns, corr_list))

# Create a dataframe to store our values
corr_df = pd.DataFrame(column_corr, columns=['Column_name', "Corr_to_price"])
corr_df.sort_values(by='Corr_to_price', ascending=False)
```

```python
# Create a function for plotting scatter plots
def Scatter_plot(X_data, y_data, column):
    """
    A function that does a scatterplot for the column specified in a given dataframe
    """
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.scatter(X_data[column], y_data, c='r')
        
    ax.set_xlabel(column)
    ax.set_ylabel(y_data.name) 
    ax.set_title(f'{column.title()} scatterplot')
    
    plt.savefig(f'/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/{column.title()} scatterplot.jpg')
    plt.show()
```

```python
# Create a for loop to loop through our function for plotting
for column in X_data.columns:
    # call the function
    Scatter_plot(X_data, y_data, column)
```

###### Correlation and Scatter plot Interpretation

The top features correlated to price are:
- sqft_living
- grade
- bathrooms
- view
- bedrooms

A bit more in-depth interpretation:
- Bedroom price is somewhat normally distributed, with peak price hitting at 6 bedrooms and then they start dipping.
- Price increases gradually in relation to bathroom count
- sqft_living has a good direct proportionality with price
- 2 floor homes generally attract a higher pricing
- Generally more views translate to higher price
- Condition is somewhat normally distributed with peak hitting at 3
- The higher the grade the higher the price in general

### Multivariate analysis

Here we are going to look at how columns correlate to each other with an exception of price.

```python
# Make dataframe of correlation values 
corrs = data.drop('price', axis=1).corr()

# Set figure and axes
fig, ax = plt.subplots(figsize=(16, 10))
fig.set_tight_layout(True)

# Plot the heatmap
sns.heatmap(
    # Data for plotting
    corrs, 
    # Axis to plot on
    ax=ax, 
    # Show correlation values
    annot=True,
    # To set a diverging color gradient
    center=0,
    # Get rid of the other half of duplicates
    mask=np.triu(np.ones_like(corrs, dtype=bool)),
    # Set name of key/legend
    cbar_kws={"label": "Correlation"}
)

# Set the figure title
ax.set_title('Correlation of all features in our data(including the target)')
plt.xticks(rotation=45)

# Save the chart 
plt.savefig('/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/Correlation_of_features.jpg')
plt.show()
```
