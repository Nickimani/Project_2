## Business Problem

Unsung real estate agency is an agency in King County that deals with clients who want to buy/sell homes.

They want to create an algorithm and use it to make a model for pricing of houses:
- The primary purpose of the model is to predict buying/selling prices of houses.
- The secondary purpose is for inferring features/atrributes that significantly affect the pricing of a house. 

By completing this, the business is at a better position to advise clients who want to buy/sell homes with factual information about homes, their costs or selling prices and exactly why their prices are as they are.

## Data Summary

The data used is for the for the year 2014 May - 2015 May. The data has no missing records or inappropriate datatypes. There are 21597 records of homes in King County, their attributes, renovation year if any, and their selling price. This is significant information to use for modelling and inferencing in line with  our business problem.

For our model:
- The target is __`price`__ which is the pricing/value of a house.
- `sqft_lot` is the area of the whole plot of land.
- `sqft_living` is the area which the house/home takes in the plot.
- `floors` are the levels in a house.
- `waterfront` is whether a house has a view of a waterfront or not.
- `bedrooms`and `bathrooms` are the number of the respective rooms in a house.
- `condition` is the overall condition of a house.
- `grade` is the grading of a house.

## Overview

I managed to find out that:
- Bedroom price is somewhat normally distributed, with peak price hitting at 6 bedrooms and then the bedrooms count starts dipping.
- Price increases gradually in relation to bathroom count
- sqft_living has a good direct proportionality with price
- 2 floor homes generally attract a higher pricing
- Generally more views translate to higher price
- Condition is somewhat normally distributed with peak hitting at 3. Meaning price increases gradually upto a condition of 3, then it starts dropping.
- The higher the grade the higher the price in general

Also, based on the final model: 

The base price of a house is usually at ~15,880,000 dollars +/- ~217,461.106 dollars. After checking the attributes for a house, appropriate changes are made to the price. For example a house having a poor grade reduces the price by ~302,900 dollars while a house having an excellent grade increases the price by ~861,900 dollars.

I am going to change the category classifications for each of these categorical columns. This will be as a way of preparing the columns for one-hot encoding.

`grade` and `renovated` are not being pre-processed although they are categorical features because their contents are binary like so it won't affect the model.

```python
# Define a function for one-hot encoding here 
def One_hot(data, column):
    """
    The function takes in a dataframe and a categorical column that requires one-hot encoding and returns 
    the dataframe with one hot encoding done on it
    """
    # Import the neccessary module
    from sklearn.preprocessing import OneHotEncoder
    
    # declare a variable for the column in the dataframe
    pross_data = data[[column]]
    # Instantiate one hot encoder with neccessary parameters
    ohe = OneHotEncoder(drop='first', categories='auto', sparse=False, handle_unknown='error')
    # fit the encoder on the data
    ohe.fit(pross_data)
    # transform the data to be processed
    en_pross_data = ohe.transform(pross_data)
    
    # create a new dataframe for eaze in adding back the encoded columns to our initial dataframe
    df_en_pross = pd.DataFrame(en_pross_data,
                               # set column names 
                               columns = [(f'{column}_' + str(x)) for x in ohe.categories_[0][1:]],
                               # set the index same as the one for our original dataframe so as to effectively merge 
                               index=data.index)
    
    # drop the column that has been encoded 
    data.drop(columns=column, inplace=True)
    # concatenate the encoded columns with our original dataframe
    merged_data = pd.concat([data, df_en_pross], axis=1)
    
    
    return merged_data
```

In order to normalize the distribution, we are going to apply log transfomation or mean standardization to these columns

```python
def min_max_scaler(data, column):
    """
    Applies min max scaling to the column of choice and drops the column while adding the new scaled column to 
    the dataframe
    """
    # store necessary values in variables
    min_ = data[column].min()
    max_ = data[column].max()
    range_ = max_ - min_
    
    scaled_column = [] 
    
    for num in data[column]:
        result = (num - min_) / range_
        scaled_column.append(result)
    # Create a new scaled columns in the dataset and drop the original one
    data['scaled_' + column] = scaled_column
    
    data.drop(columns=column, inplace=True)
    
    return data
```

```python
def log_transfromer(data, column):
    """
    Applies log transformation to the column of choice and drops the column while adding the new scaled column to 
    the dataframe
    """
    # store necessary values in variables
    tran_column = np.log(data[column])
    
    # Create a new transformed column in the dataset and drop the original one
    data['transformed_' + column] = tran_column
    
    data.drop(columns=column, inplace=True)
    
    return data
```

### Check how the model performs according to the linear regression assumptions

#### Investigating linearity

Now let's check whether the linearity assumption will hold for our model.

```python
fig, ax = plt.subplots()

perfect_line = np.arange(y_test.min(), y_test.max())
ax.plot(perfect_line, linestyle="--", color="orange", label="Perfect Fit")
ax.scatter(y_test, pred, alpha=0.5)
ax.set_xlabel("Actual Price")
ax.set_ylabel("Predicted Price")
plt.savefig('/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/Linearity assumption.jpg')
ax.legend();
```

The relationship between our predicted price and actual price is somewhat linear and we can consider ourselves as not violating the linearity assumption.

#### Investigating Normality

Now let's check whether the normality assumption will hold for our model.

```python
import scipy.stats as stats

residuals = (y_test - pred)
sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.savefig('/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/Normality assumption.jpg')
plt.show()
```

Except for the outliers on the top half, we can say that our model is in-line with the normality assumption.

#### Investigating Multicollinearity 

The rule for VIF is that 5 and above is an indicator of strong multicollinearity. Let's see how our model performs here.

```python
# The module imported is great for checking multicollinearity in a model.
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]

pd.Series(vif, index=X_train.columns, name="Variance Inflation Factor")
```

The features that have strong multicollinearity are the had the highest correlation to our target back when we checked. 

For transformed_sqft_living and transformed_sqft_living_sq, they have very strong multicollinearity because the latter is a squared version of the former. So the high values are to be expected.

Despite these highlights, the model performs not so bad when checking the multicollinearity assumption.

#### Investigating Homoscedasticity

Now let's check whether the model errors are homoscedastic or not.

```python
fig, ax = plt.subplots()

ax.scatter(pred, residuals, alpha=0.5)
ax.plot(pred, [0 for i in range(len(X_test))])
ax.set_xlabel("Predicted Value")
ax.set_ylabel("Actual - Predicted Value")
plt.savefig('/home/nick/Documents/Flatiron/Phase 2/Project_2/Images/Homoscedasticity assumption.jpg')
```

Though it is not the best division of errors, it is not bad either.

The distribution of errors in our model can be said to be homoscedastic
